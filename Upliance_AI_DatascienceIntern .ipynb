### Handling Missing Values
 In Order_details table Rating Column contains some null values

```python
order_df['Rating'].fillna(order_df['Rating'].mean(),inplace=True)
```

```python
datasets=[order_df,user_df,cooking_df]
for i in range(len(datasets)):
    print(datasets[i].isnull().sum())
```

### Merging Tables

```python
merged_data=pd.merge(user_df,cooking_df,on="User ID",how='inner')
final_df=pd.merge(merged_data,order_df,on="User ID",how='inner')
```

## Demographics and User Behavior

### Total Orders by Age Group

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=final_df, x='Age', y='Total Orders', palette='coolwarm')
plt.title('Total Orders by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Orders')
plt.show()
```

### Which region has highest number of orders

```python
final_df.groupby(['Location']).sum()["Total Orders"].plot.pie()
```

### Cooking Sessions vs Orders

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=final_df, x='Duration (mins)', y='Total Orders', hue='Session Rating', palette='viridis')
plt.title('Duration vs. Total Orders (Color by Session Rating)')
plt.xlabel('Duration (mins)')
plt.ylabel('Total Orders')
plt.legend(title='Session Rating')
plt.show()
```
